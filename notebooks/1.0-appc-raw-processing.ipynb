Transform csv data for psychoactive substance abuse and population into parquet for repository upload

```python
import pandas as pd
import numpy as np
```

```python
#read csv for psychoactive substance abuse
psycho_substance=pd.read_csv("/home/ange/DS_projects/psycho_substance_abuse/data/descargable_vespa_gral.csv",sep=';')
#read csv for demographic data
population=pd.read_csv("/home/ange/DS_projects/psycho_substance_abuse/data/osb_demografia-poblacion-localidad.csv", 
                       sep=';',encoding='latin-1')
```

```python
#remove nulls from psychoactive dataframe and convert 'CASOS' column to numeric
psycho_substance.replace({'CASOS':' '},np.nan,inplace=True)
psycho_substance['CASOS']=psycho_substance['CASOS'].apply(pd.to_numeric)
```

```python
#Transform to parquet
psycho_substance.to_parquet('/home/ange/DS_projects/psycho_substance_abuse/data/raw/psycho_substance.parquet.gzip',
                             compression='gzip',engine='pyarrow')
population.to_parquet('/home/ange/DS_projects/psycho_substance_abuse/data/raw/population.parquet.gzip',
                             compression='gzip',engine='pyarrow')
```
